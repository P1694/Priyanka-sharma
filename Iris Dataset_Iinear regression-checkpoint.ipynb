##### Modeling

Train Test Split

```python
#variable
x= iris_df
y=target_df
```

```python
# Splitting the Dataset 
train_x, test_x, train_y, test_y=train_test_split(x,y, test_size=.33, random_state=0)
```

Linear Regression 

```python
from sklearn.linear_model import LinearRegression
LR= LinearRegression()
```

```python
LR.fit(train_x, train_y)
```

```python
LR.score(train_x,train_y)
```

```python
LR_Pred=LR.predict(test_x)
LR_Pred
```

```python
df=pd.DataFrame(LR_Pred,columns=['Predicted'])
actuals=test_y
actuals.reset_index(drop=True, inplace=True)
```

```python
df3=pd.concat([df,actuals],axis=1)
```

Error testing

```python
print('Mean Absolute Error:', mean_absolute_error(test_y, LR_Pred))
print('Mean Squared Error:', mean_squared_error(test_y, LR_Pred))
print('Mean Root Squared Error:', np.sqrt(mean_squared_error(test_y, LR_Pred)))
```
