```python
from sklearn.datasets import load_boston
Bs=load_boston()
Bs
```

```python
Bn_dat=pd.DataFrame(data=Bs.data, columns=Bs.feature_names)
Bn_target=pd.DataFrame(data=Bs.target, columns=['Target'])
```

Treatment of outlier

```python
Bn_dat[Bn_dat["CRIM"]>40]=40
Bn_dat[Bn_dat["ZN"]>40]=40
Bn_dat[Bn_dat["RM"]>8]=8
Bn_dat[Bn_dat["DIS"]>10]=10
Bn_dat[Bn_dat["LSTAT"]>30]=30
```

Labling the data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
L=LabelEncoder()
for i in Bn_dat:
    if Bn_dat[i].dtypes== 'object':
        Bn_dat[i]=L.fit_transform(Bn_dat[i].astype(str))
```

Modeling

Train Test Split

```python
#variable
x= Bn_dat
y=Bn_target
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting the Dataset 
train_x, test_x, train_y, test_y=train_test_split(x,y, test_size=.33, random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
RFR=RandomForestRegressor()
```

```python
RFR.fit(train_x,train_y)
```

```python
RFR.score(train_x,train_y)
```

```python
RFR_Pred=RFR.predict(test_x)
RFR_Pred
```

```python
RFR_Pred=pd.DataFrame(RFR_Pred,columns=['Predicted'])
actuals=test_y
actuals.reset_index(drop=True, inplace=True)
df3=pd.concat([RFR_Pred,actuals],axis=1)
df3
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print('Mean Absolute Error:', mean_absolute_error(test_y, RFR_Pred))
print('Mean Squared Error:', mean_squared_error(test_y, RFR_Pred))
print('Mean Root Squared Error:', np.sqrt(mean_squared_error(test_y, RFR_Pred)))
```
